```python
import numpy as np
import pandas as pd
import ast
import random
from top2vec import Top2Vec
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/NLP/Handin/Data/df_reviews_final_binary.csv')
```

```python
df.shape
```

```python
random.seed(42)
```

```python
model = Top2Vec(df['review_comments'].tolist())
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
print(topic_sizes[:20])
```

```python
print(topic_nums)
```

```python
topic_words, word_scores, topic_nums = model.get_topics(20)
```

```python
for words, scores, num in zip(topic_words, word_scores, topic_nums):
    print(num)
    print(f'Words: {words}')
```

```python
model.generate_topic_wordcloud(0, background_color='white')
```

```python
model.generate_topic_wordcloud(1, background_color='white')
```

```python
topic_words, word_scores, topic_scores, topic_nums = model.search_topics(keywords=["price"], num_topics=3)
```

```python
for words, scores, num in zip(topic_words, word_scores, topic_nums):
    print(num)
    print(f'Words: {words}')
```
